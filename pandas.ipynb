```python
s=str()
n=int(input('enetr n: '))
for i in range(n):
    name=input('name: ')
    phone=int(input('number: '))
    
    s=s+f'{name},{phone}\n'
print(s)

b=open('phone.csv','a')
b.write(s)
b.close()
```

```python
s=str()
n=int(input('enetr n: '))
for i in range(n):
    name=input('name: ')
    age=int(input('enter age: '))
    add=input('enetr address: ')
    phone=int(input('number: '))
    
    s=s+f'{name},{age},{add},{phone}\n'
print(s)

b=open('phone.csv','w')
b.write(s)
b.close()
```
